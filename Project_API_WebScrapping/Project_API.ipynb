```python
import requests,time
from bs4 import BeautifulSoup
import numpy as np

def clean(result):
    return [i.text.strip().split("\n") for i in result]

def convert_frame(clean):
    return pd.DataFrame(clean)

def concatenar_frame(f1,f2,f3):
    return pd.concat([f1,f2,f3],axis=1)

def eliminar_vacios(data):
    return data.dropna(how='all')

big_concat=pd.DataFrame()

for k in range(1,21):
    soup = (BeautifulSoup(requests.get(f'https://www.packtpub.com/all-products/all-books?released=Available&page={k}').content,"lxml")) 
    print(requests.get(f'https://www.packtpub.com/all-products/all-books?released=Available&page={k}').status_code)
    name= soup.find_all('a',class_='product-item-link')
    date_publication= soup.find_all('div',class_='date-of-publication')
    price= soup.find_all('span',class_='price')
    name_clean,date_publication_clean,price_clean =clean(name),clean(date_publication),clean(price)
    name_frame,date_frame,price_frame =convert_frame(name_clean),convert_frame(date_publication_clean),convert_frame(price_clean)
    allconcat=concatenar_frame(name_frame,date_frame,price_frame)
    allconcatna=eliminar_vacios(allconcat)
    print(allconcatna)
    time.sleep(5)
    big_concat=big_concat.append(allconcatna)

big_concat
```
