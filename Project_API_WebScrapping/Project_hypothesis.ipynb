```python
#Adquisition of inequality data 
from bs4 import BeautifulSoup
import pandas as pd
import json
import requests
import numpy as np 
import pandas as pd
from pandas.io.json import json_normalize

def trans_(data):
    return data.transpose()
def dataframe(data):
    return pd.DataFrame(data)

def selecciona_delito(url):
    response=requests.get(url)
    results=response.json()
    return results

def elimina_columnas(data,x):
    return data.drop(x,axis=1)

def flattened_data(results):
    flattened_data = json_normalize(results)
    flattened_data1 = json_normalize(flattened_data.records[0])
    data=dataframe(flattened_data1)
    return data
    
def maxi(results,column):
    return results[column].value_counts().nlargest(1)


bash_columns=['datasetid','fields.categoria_delito','fields.fecha_hechos','recordid','record_timestamp','fields.geopoint','fields.fecha_inicio','fields.agencia','fields.fiscalia','fields.unidad_investigacion','fields.longitud','geometry.type','geometry.coordinates','fields.latitud','fields.ao_inicio','fields.delito','fields.calle_hechos2','fields.mes_inicio','fields.mes_hechos','fields.calle_hechos','fields.colonia_hechos','fields.ao_hechos']    

#Robo a trausente con y sin violencia
results=elimina_columnas(flattened_data(selecciona_delito('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=carpetas-de-investigacion-pgj-cdmx&rows=10000&facet=ao_hechos&facet=mes_hechos&facet=delito&facet=categoria_delito&facet=fiscalia&facet=colonia_hechos&facet=alcaldia_hechos&facet=ao_inicio&refine.ao_hechos=2018&refine.categoria_delito=ROBO+A+TRANSEUNTE+EN+V%C3%8DA+P%C3%9ABLICA+CON+Y+SIN+VIOLENCIA')),
                bash_columns)
print('Robo a transeúnte con o sin violencia 2018: ', maxi(results,'fields.alcaldia_hechos'))


#Robo a trausente - celular
results=elimina_columnas(flattened_data(selecciona_delito('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=carpetas-de-investigacion-pgj-cdmx&rows=10000&facet=ao_hechos&facet=mes_hechos&facet=delito&facet=categoria_delito&facet=fiscalia&facet=colonia_hechos&facet=alcaldia_hechos&facet=ao_inicio&refine.ao_hechos=2018&refine.delito=ROBO+A+TRANSEUNTE+DE+CELULAR+CON+VIOLENCIA')),
                bash_columns)
print('Robo a transeúnte celular: ', maxi(results,'fields.alcaldia_hechos'))


#Violencia familiar
results=elimina_columnas(flattened_data(selecciona_delito('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=carpetas-de-investigacion-pgj-cdmx&rows=10000&facet=ao_hechos&facet=mes_hechos&facet=delito&facet=categoria_delito&facet=fiscalia&facet=colonia_hechos&facet=alcaldia_hechos&facet=ao_inicio&refine.delito=VIOLENCIA+FAMILIAR&refine.ao_hechos=2018')),
                bash_columns)
print('Violencia familiar: ', maxi(results,'fields.alcaldia_hechos'))

#Negocio sin violencia
results=elimina_columnas(flattened_data(selecciona_delito('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=carpetas-de-investigacion-pgj-cdmx&rows=10000&facet=ao_hechos&facet=mes_hechos&facet=delito&facet=categoria_delito&facet=fiscalia&facet=colonia_hechos&facet=alcaldia_hechos&facet=ao_inicio&refine.ao_hechos=2018&refine.delito=ROBO+A+NEGOCIO+SIN+VIOLENCIA')),
                bash_columns)
print('Robo a negocio sin violencia: ', maxi(results,'fields.alcaldia_hechos'))

#Fraude
results=elimina_columnas(flattened_data(selecciona_delito('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=carpetas-de-investigacion-pgj-cdmx&rows=10000&facet=ao_hechos&facet=mes_hechos&facet=delito&facet=categoria_delito&facet=fiscalia&facet=colonia_hechos&facet=alcaldia_hechos&facet=ao_inicio&refine.ao_hechos=2018&refine.delito=FRAUDE')),
                         bash_columns)
print('Fraude : ', maxi(results,'fields.alcaldia_hechos'))

#Amenazas
results=elimina_columnas(flattened_data(selecciona_delito('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=carpetas-de-investigacion-pgj-cdmx&rows=10000&facet=ao_hechos&facet=mes_hechos&facet=delito&facet=categoria_delito&facet=fiscalia&facet=colonia_hechos&facet=alcaldia_hechos&facet=ao_inicio&refine.delito=AMENAZAS',)),
                         bash_columns)
print('Amenazas : ', maxi(results,'fields.alcaldia_hechos'))

# Impresión deLas 2 delegaciones con más indices delectivos

data_desigualdad = pd.read_csv('Desigualdad_2018_CDMX.csv')
data=dataframe(data_desigualdad)
data_Cua=trans_(data.loc[data['Alcaldía'] == 'Cuauhtémoc']).rename(columns={0: "Datos"})
data_Itzta=trans_(data.loc[data['Alcaldía']=='Iztapalapa']).rename(columns={8: "Datos"})
data=pd.concat([data_Cua,data_Itzta],axis=1)
data
```
